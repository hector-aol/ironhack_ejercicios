```python
alturas = [34,36,37,41,21,51,34,32,52]
```

```python
alturas = sorted(alturas)
n = len(alturas)

def mediana(l):
    m = len(l)
    mitad = m //2

    if m%2 == 0:
        return (l[mitad-1] + l[mitad])/2
    else:
        return l[mitad]
    
Q2 = mediana(alturas)
if n%2 == 0:
    a_i =alturas[:n//2]
    a_s =alturas[n//2:]
else:
    a_i =alturas[:n//2]
    a_s =alturas[n//2+1:]

Q1 , Q3 = mediana(a_i) , mediana(a_s)

maximo = max(alturas)
minimo = min(alturas)
media = sum(alturas)/n

print(f"datos = {alturas}")

print(f"Min:{minimo}")
print(f"Q1:{Q1}")
print(f"Q2=Mediana:{Q2}")
print(f"Q3:{Q3}")
print(f"Max:{maximo}")

print(f"Media:{media}")

RI = Q3-Q1

INF = Q1 -1.5*RI
SUP = Q3 + 1.5*RI

print(f"RI = {RI}  inf = {INF}   sup = {SUP}")

OUT_INF , OUT_SUP = [k for k in alturas if k < INF] , [k for k in alturas if k > SUP]

print(f"Outliers inferiores = {", ".join(OUT_INF)}   -    Outliers superiores = {", ".join(OUT_SUP)}")

plt.boxplot(alturas)
plt.show()
```

```python
import numpy as np
q1 = np.percentile(alturas,25)
q3 = np.percentile(alturas,75)
ri = q3-q1

inf = q1 -1.5*ri
sup = q3 + 1.5*ri

print(f"RI = {ri}  inf = {inf}   sup = {sup}")

out_inf , out_sup = [k for k in alturas if k < inf] , [k for k in alturas if k > sup]

print(f"Outliers inferiores = {out_inf}   -    Outliers superiores = {out_sup}")
```

```python
import random
import seaborn as sns


pob1 = [random.gauss(30,3) for _ in range(100)]
pob2 = [random.gauss(70,3) for _ in range(100)]

pob = pob1 + pob2

random.shuffle(pob)

sns.kdeplot(pob , fill=True , color= 'skyblue')

plt.show()
```

# sns vs plt

```python
x, y = [1, 2, 3, 4], [1, 2, 7, 0.5]

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

sns.lineplot(x=x, y=y, ax=axes[0], color='blue')
axes[0].fill_between(x, y, color='lightblue', alpha=0.8)
axes[0].set_title("Seaborn (lnea + rea)")

axes[1].fill_between(x, y, color='orange', alpha=0.7)
axes[1].set_title("Matplotlib - fill_between")

plt.tight_layout()
plt.show()
```

```python
x = np.random.normal(5, 1.6, 1500)


fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.histplot(x, ax=axes[0], color='skyblue')
axes[0].set_title("Seaborn - histplot")


axes[1].hist(x, bins=30, color='salmon', edgecolor='black')
axes[1].set_title("Matplotlib - hist")

plt.tight_layout()
plt.show()
```
