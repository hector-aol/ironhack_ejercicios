1. Modelos Supervisados (Supervised Learning)
Concepto: Aprenden a partir de ejemplos etiquetados. Cada dato de entrenamiento incluye la entrada (características) y la salida correcta esperada (etiqueta).

Cómo funciona: El algoritmo busca patrones que mapeen las entradas a las salidas conocidas. Una vez entrenado, predice etiquetas para datos nuevos nunca vistos.

Objetivo: Predecir resultados futuros o clasificar datos basándose en ejemplos históricos.

Tipos principales:

Clasificación: Predecir una categoría/clase (e.g., spam/no spam, diagnóstico médico, reconocimiento de imágenes).

Regresión: Predecir un valor numérico continuo (e.g., precio de una casa, temperatura futura, ventas estimadas).

Ejemplos de algoritmos: Regresión lineal, Regresión logística, Máquinas de Soporte Vectorial (SVM), Árboles de decisión, Random Forest, Redes Neuronales (para clasificación/regresión).

Analogía: Como un estudiante que aprende con un libro de texto que tiene las respuestas correctas al final. Compara sus respuestas con las correctas y ajusta su método.

2. Modelos No Supervisados (Unsupervised Learning)
Concepto: Aprenden a partir de datos sin etiquetar. Solo se proporcionan las entradas (características); no hay salidas "correctas" conocidas.

Cómo funciona: El algoritmo explora los datos para descubrir estructura inherente, patrones ocultos, agrupaciones naturales o relaciones entre variables.

Objetivo: Encontrar organización o significado en datos complejos, reducir dimensiones, identificar grupos similares.

Tipos principales:

Clustering (Agrupamiento): Agrupar datos similares (e.g., segmentación de clientes, agrupación de documentos por tema).

Reducción de dimensionalidad: Simplificar datos manteniendo su esencia (e.g., PCA, t-SNE).

Detección de anomalías: Identificar puntos de datos inusuales o atípicos.

Reglas de asociación: Descubrir relaciones entre variables (e.g., "los clientes que compran X también suelen comprar Y").

Ejemplos de algoritmos: K-Means, DBSCAN, Análisis de Componentes Principales (PCA), Autoencoders (no supervisados).

Analogía: Como explorar una biblioteca desconocida sin índice. Agrupas libros por temas similares basándote solo en su contenido y portada, sin saber de antemano las categorías.

3. Modelos por Refuerzo (Reinforcement Learning - RL)
Concepto: Un agente aprende a tomar decisiones secuenciales dentro de un entorno para maximizar una recompensa acumulada a lo largo del tiempo.

Cómo funciona:

El agente observa el estado del entorno.

Realiza una acción.

Recibe una recompensa (o castigo) del entorno.

Transiciona a un nuevo estado.

Aprende una política (estrategia) que le dice qué acción tomar en cada estado para maximizar las recompensas futuras.

Objetivo: Aprender la secuencia óptima de acciones para lograr un objetivo complejo a largo plazo en un entorno dinámico e incierto.

Características clave: No hay datos etiquetados, solo señales de recompensa (a menudo escasas o retardadas). Requiere exploración (probar cosas nuevas) y explotación (usar lo que sabe que funciona).

Aplicaciones: Robótica (control de movimiento), Juegos (AlphaGo, Dota 2), Gestión de recursos (inventario, energía), Sistemas de recomendación interactivos, Vehículos autónomos (ciertos aspectos).

Ejemplos de algoritmos: Q-Learning, SARSA, Deep Q-Networks (DQN), Policy Gradients, Proximal Policy Optimization (PPO).

Analogía: Como entrenar a un perro con premios (recompensas) o regañinas (castigos). El perro (agente) prueba diferentes acciones (sentarse, dar la pata) en diferentes situaciones (entorno) y aprende qué acciones le dan más premios a largo plazo.

4. Modelos Auto-supervisados (Self-Supervised Learning)
Concepto: Un subconjunto del aprendizaje no supervisado donde los datos generan sus propias "etiquetas" automáticamente a partir de su estructura inherente, sin necesidad de etiquetado humano.

Cómo funciona: Se diseñan tareas pretexto (pretext tasks) que el modelo debe resolver usando solo partes no etiquetadas de los datos. Resolver estas tareas ayuda al modelo a aprender representaciones internas ricas y útiles de los datos.

Objetivo: Aprender representaciones generales de alta calidad de los datos (imágenes, texto, sonido) que luego se pueden usar para tareas específicas posteriores (supervisadas o no supervisadas) con menos datos etiquetados (transfer learning).

Ejemplos de tareas pretexto:

Texto: Predecir una palabra oculta en una oración (e.g., BERT, GPT).

Imágenes: Reconstruir partes de una imagen ocultas ("inpainting"), predecir la rotación aplicada a una imagen, predecir la relación espacial entre parches de imagen.

Video: Predecir el orden correcto de tramas desordenadas, predecir tramas futuras/pasadas.

Ventaja clave: Permite aprovechar enormes cantidades de datos no etiquetados (muy abundantes) para aprender características poderosas, reduciendo la dependencia de costosos datos etiquetados.

Relación con otros modelos: Es una forma de entrenar modelos (usualmente redes neuronales profundas) utilizando datos no etiquetados. El modelo entrenado auto-supervisado luego se puede:

Usar directamente para algunas tareas (e.g., generación de texto con GPT).

Afinar (fine-tuning) con pocos datos etiquetados para una tarea específica (e.g., clasificación de sentimientos usando BERT).

Ejemplos prominentes: BERT, GPT (y toda la familia GPT-3, GPT-4, etc.), DALL-E (parte del entrenamiento), modelos de lenguaje grande (LLMs) en general.

Analogía: Como aprender a armar un rompecabezas. Las piezas mismas (los datos) te dan las pistas sobre cómo encajan (las "etiquetas" auto-generadas). Una vez que aprendes a armar rompecabezas (representación aprendida), puedes armar nuevos rompecabezas similares (tareas específicas) más fácilmente.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

np.random.seed(42)
x = np.sort(5 * np.random.rand(80,1), axis = 0)
y = 2 * x.squeeze() + 3 + np.random.randn(80)*1.5
```

```python
plt.scatter(x, y, color='blue', label='datos')
plt.legend()
plt.show()
```

```python
reg_1 = LinearRegression()
reg_1.fit(x,y)

pred_reg_1 = reg_1.predict(x)

plt.plot(x, pred_reg_1, color='red', label='Reg. Lineal')
```

```python
reg_1 = LinearRegression()
reg_1.fit(x,y)

pred_reg_1 = reg_1.predict(x)

plt.plot(x, pred_reg_1, color='red', label='Reg. Lineal')
plt.scatter(x, y, color='blue', label='datos')
plt.legend()
plt.show()
```
